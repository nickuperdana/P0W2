## Task 1
  
How many tables are in the austin_bikeshare dataset?

### Answer
  
To run the query against a project other than your default project, add the project ID to the dataset in the following format: `project_id`.dataset.INFORMATION_SCHEMA.view;  

for example,  
`myproject`.mydataset.INFORMATION_SCHEMA.TABLES.

```python
# put your query here
query = """
SELECT *
FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

## Task 2
  
How many columns in the bikeshare_stations table have STRING data?

### Answer
  
Add the project ID to the dataset in the following format: `project_id`.dataset.INFORMATION_SCHEMA.view;  

for example, `bigquery-public-data`.census_bureau_usa.INFORMATION_SCHEMA.TABLES.

```python
# put your query here
query = """
  SELECT
    *
  FROM
    `bigquery-public-data`.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS
  WHERE
    table_name = 'bikeshare_stations' AND data_type = 'STRING';
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

# Question 2  



## Task 1  

Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes

### Answer

```python
# put your query here
query = """
  SELECT *
  FROM
    `bigquery-public-data`.austin_bikeshare.bikeshare_trips
  WHERE
    bike_id iS NOT NULL AND
    duration_minutes > 1000;
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

## Task 2  

Filter trips where the trip_id is "Walk Up" and the end_station_name is "Stolen".

### Answer

```python
# put your query here
query = """
  SELECT
    *
  FROM
    `bigquery-public-data`.austin_bikeshare.bikeshare_trips
  WHERE
    subscriber_type = 'Walk Up'
    AND end_station_name = 'Stolen';
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```
