````python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()

project_id = "data-science-408704"
client = bigquery.Client(project=project_id)
````

````python
df = client.query('''
SELECT *
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
''').to_dataframe()

df
````

#####1. Berapa banyak label unique dari bike_type?

````python
# working area
df_1 = client.query('''
SELECT COUNT(DISTINCT bike_type) AS total_bike_type
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
''').to_dataframe()

df_1
````

Insight: Ada dua macam sepeda yang direntalkan oleh layanan Austin Bikeshare

#####2. Apa saja nama label unique dari tipe sepeda?

````python
# working area
df_2 = client.query('''
SELECT DISTINCT bike_type AS list_of_bike_type
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
''').to_dataframe()

df_2
````

Insight: Ada dua macam sepeda yang direntalkan, yakni sepeda tipe klasik, dan sepeda tipe elektrik

#####3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

````python
# working area
df_3 = client.query('''
SELECT COUNT(start_time)
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
WHERE EXTRACT(YEAR FROM start_time) = 2016
OR EXTRACT(YEAR FROM start_time) = 2017
''').to_dataframe()

df_3

# df3_2016_2017 = df_3[(df_3['start_time'].dt.year == 2016) | (df_3['start_time'].dt.year == 2017)]

# df3_2016_2017.start_time.count()
````

Insight: ada 7993 kali peminjaman sepeda pada tahun 2016 dan 2017

#####4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

````python
# working area
df_4 = client.query('''
SELECT start_station_name,
COUNT(start_station_name) AS total
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
WHERE EXTRACT(YEAR FROM start_time) = 2016
GROUP BY start_station_name
ORDER BY total DESC
LIMIT 1;
''').to_dataframe()

df_4
````

Insight: Bullock Museum @ Congress & MLK merupakan stasiun keberangkatan yang paling sering dikunjungi oleh pengguna sewa sepeda pada 2016, dengan total kunjungan sebanyak 394 kali.

#####5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

````python
# working area
df_5 = client.query('''
SELECT MIN(duration_minutes) AS minimum_trip_duration,
  MAX(duration_minutes) AS maximum_trip_duration,
  AVG(duration_minutes) AS mean_trip_duration
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
''').to_dataframe()

df_5
````

Insight: Rental sepeda paling singkat pada keseluruhan rekaman data ialah 2 menit. Rental sepeda paling lama ialah selama 2218 menit. Rata-rata durasi rental sepeda secara keseluruhan ialah 35 menit.

#####6. Berapa rata-rata durasi perjalanan dari subscriber_type `Student Membership` dan `Local31`?

````python
# working area
df_6 = client.query('''
SELECT AVG(duration_minutes)
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
WHERE subscriber_type = 'Student Membership' OR subscriber_type = 'Local31';
''').to_dataframe()

df_6
````

Insight: Tidak ada rata-rata lama durasi peminjaman sepeda oleh mereka yang memegang tipe subscription 'Student Membership' dan 'Local31'.

#####7. Tampilkan entry data yang memiliki `duration_time` diatas 4.25 jam. (*Gunakan subquery untuk menjawab pertanyaan ini dengan meng-convert duration_time dari menit ke hour*).

````python
# working area
df_6 = client.query('''
SELECT duration_minutes
FROM `hacktiv8-ftds-challenge.practice.bike_rent`
WHERE duration_minutes/60 > 4.25
# WHERE 4.25 < (
#   SELECT duration_minutes/60
#   FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`

''').to_dataframe()
df_6
````

#####2. Buat kolom baru bernama `quarter`.
  
  Lakukan ekstraksi waktu (quarter dalam tahun) berdasarkan kolom `start_time` dengan ketentuan berikut:

<div align="center">

  |Keterangan Waktu|Bulan|
  |--|--|
  |Q1|Januari, Februari, Maret|
  |Q2|April, Mei, Juni|
  |Q3|Juli, Agustus, September|
  |Q4|Oktober, November, Desember|

</div>

>**Notes:**
>- Dalam menjawab soal Pandas no. 2, **tidak boleh** menggunakan ekstraksi waktu `Series.dt.quarter`.
- Lakukan `looping` dan `conditional` dengan memanfaatkan ekstraksi bulan `Series.dt.month` terhadap kolom `start_time`, pastikan tipe datanya ialah `datetime` agar dapat dilakukan ekstraksi.
- Simpan hasil looping ke dalam suatu variabel, dan buat kolom baru berisikan variabel sebelumnya.

````python
# working area
# cleaning datanya dulu sebelum jawab nomor 2

quarterColumn = []

dfWithColumnQuarter = df.copy()

for entry in dfWithColumnQuarter.start_time.dt.month:
  if entry <= 3:
    quarterColumn.append('Q1')
  elif entry <= 6:
    quarterColumn.append('Q2')
  elif entry <= 9:
    quarterColumn.append('Q3')
  else:
    quarterColumn.append('Q4')

# dfWithColumnQuarter['quarter'] = dfWithColumnQuarter.start_time.dt.month

dfWithColumnQuarter['quarter'] = quarterColumn

dfWithColumnQuarter
````

#####3. Tampilkan data perjalanan dari tipe member `Local31` dengan durasi perjalanan diatas 1 jam.
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*

````python
# working area

dfLocal31 = df.copy()

dfLocal31 = dfLocal31[(dfLocal31['subscriber_type'] == 'Local31') & (dfLocal31['duration_minutes'] > 60)]

dfLocal31.sort_values(by='duration_minutes')
````

Insight: Tidak ada data yang memuat subscriber_type 'Local31' dengan data durasi perjalanan lebih lama dari satu jam.

#####4. Berapa rata-rata durasi perjalanan dari tipe pelanggan `Walk Up` dengan tipe sepeda `classic` pada bulan `Maret` dari data yang diambil?
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*

````python
# working area

dfWalkUpClassicinMarch = df.copy()

dfWalkUpClassicinMarch = dfWalkUpClassicinMarch[(dfWalkUpClassicinMarch['subscriber_type'] == 'Walk Up') & (dfWalkUpClassicinMarch['bike_type'] == 'classic') & (dfWalkUpClassicinMarch.start_time.dt.month == 3)]

dfWalkUpClassicinMarch.duration_minutes.mean()
````

Insight: Rata-rata durasi perjalanan dari tipe pelanggan 'Walk Up' dengan tipe sepeda 'classic' khususnya yang muncul setiap bulan Maret ialah 38 menit.
