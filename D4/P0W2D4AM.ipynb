```python
# urutkan seluruh tabel berdasarkan namanya, defaultnya secara ascending

spotify_df.sort_values(by='name')
```

```python
# Urutkan secara descending
spotify_df.sort_values(by='name', ascending=False)
```

```python
# urutkan tabel 'name', kemudian setelahnya diurutkan berdasarkan 'daily_rank'. keduanya secara descending

spotify_df.sort_values(by=['name', 'daily_rank'], ascending=False)
```

```python
# urutkan tabel 'name' secara ASCENDING, kemudian setelahnya urutkan kembali hasilnya berdasarkan 'daily_rank' berdasarkan DESCENDING karena ascending=False untuk si kolom 'daily_rank'.

spotify_df.sort_values(by=['name', 'daily_rank'], ascending=[True, False])
```

```python
# urutkan tabel 'name' secara ASCENDING, kemudian setelahnya urutkan kembali hasilnya berdasarkan 'daily_rank' berdasarkan DESCENDING karena ascending=False untuk si kolom 'daily_rank'.
# Untuk merapikan indeksnya lagi setelah melakukan sorting data, gunakan ignore_index

spotify_df.sort_values(by=['name', 'daily_rank'], ascending=[True, False], ignore_index=True)
```
