## Handling Missing Value

### Check Null Data

```python
df_spotify.info()
```

dilihat dari info, dapat kita ketahui bahwa kolom name, artists, album_name, album_release_date terdapat 1 null data, dan table country terdapat 400 null data

```python
df_spotify.isnull()
```

terlihat dari hasil isnull, 5 data terakhir pada kolom country adalah null data. berikut kita perlu menghitung jumlah null data pada kolom name, artists, album_name, album_release_date dan country.

```python
print(f'jumlah null data pada kolom name: {df_spotify.name.isnull().sum()}')
print(f'jumlah null data pada kolom artists: {df_spotify.artists.isnull().sum()}')
print(f'jumlah null data pada kolom album_name: {df_spotify.album_name.isnull().sum()}')
print(f'jumlah null data pada kolom album_release_date: {df_spotify.album_release_date.isnull().sum()}')
print(f'jumlah null data pada kolom country: {df_spotify.country.isnull().sum()}')
```

dari hasil pengecekan, diketahui benar bahwa null data pada kolom name, artists, album_name, album_release_date berjumlah 1 dan kolom country berjumlah 400 null data

### dropna

```python
# menyalin data spotify ke dalam df_drop
df_drop = df_spotify.copy()
```

```python
# menghapus seluruh null data dan kemudian index direset
df_drop = df_drop.dropna().reset_index()
```

```python
df_drop.info()
```

terlihat bahwa null data sudah dibersihkan dengan masing-masing kolom memiliki jumlah non-null data sama dengan jumlah baris yang dimiliki seluruh data

### fillna with imputation

```python
# menyalin data df_spotify ke df_fill
df_fill = df_spotify.copy()
```

```python
df_fill.info()
```

```python
df_fill.name.fillna('tanpa judul lagu', inplace=True)
df_fill.artists.fillna('tanpa penyanyi', inplace=True)
df_fill.country.fillna('tanpa negara', inplace=True)
df_fill.country.fillna('tanpa negara', inplace=True)
df_fill.album_name.fillna('tanpa album', inplace=True)
df_fill.album_release_date.fillna('tanpa tanggal rilis album', inplace=True)
```

```python
df_fill.info()
```

setelah dilakukan fillna, sekarang seluruh memiliki data yang lengkap

## Data Handling

```python
df_handling = df_spotify.copy()
```

```python
df_handling = df_handling.loc[:, 'spotify_id' : 'popularity']
```

```python
df_handling = df_handling.dropna().reset_index()
```

```python
df_handling.info()
```

```python
# melakukan pengubahan data kolom name
# dengan mengconvert data dari masing-masing element menjadi string
# kemudian memanggil fungsi strip yang ada pada tipe data str
# strip berfungsi untuk menghapus extra space
df_handling['name'] = df_handling['name'].str.strip()
```

```python
df_handling['name'] = df_handling['name'].str.replace('(', '')
df_handling['name'] = df_handling['name'].str.replace(')', '')
df_handling['name'] = df_handling['name'].str.replace('-', '')
```

```python
df_handling.head(15)
```

```python
df_handling['snapshot_date'] = pd.to_datetime(df_handling['snapshot_date'])
```

```python
df_handling.info()
```

```python
df_handling['snapshot_year'] = df_handling['snapshot_date'].dt.year
```

```python
df_handling
```

```python
df_handling['snapshot_month'] = df_handling['snapshot_date'].dt.month
```

```python
df_handling
```

```python
df_handling['snapshot_month'].value_counts()
```

```python
df_handling.to_csv('hasil_final.csv')
```
