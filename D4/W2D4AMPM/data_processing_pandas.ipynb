# Data Processing using Pandas

Pertama-tama initial dataframe terlebih dahulu

```python
import pandas as pd

pandacsv = pd.read_csv('Baby Fashion.csv')
pandacsv
```

## Missing Value

### How to check missing value

```python
pandacsv.isnull()
```

Bisa dilihat bahwa resultnya adalah True/False yang dimana True artinya value tersebut ialah null dan False artinya value tersebut non null

Kita juga bisa mengecek baris apa aja dan value apa aja yang null dengan cara berikut:

```python
null_mask = pandacsv.isnull().any(axis=1)
null_rows = pandacsv[null_mask]

null_rows
```

Terlihat bahwa ada NaN value pada data yang kita miliki

### How to drop missing value

dalam proses cleaning data, salah satu cara yang bisa kita terapkan ialah menghapus data yang kosong, untuk melakukannya kita bisa melakukan cara berikut:

```python
pandacsv_without_null = pd.read_csv('Baby Fashion.csv')

pandacsv_without_null = pandacsv_without_null.dropna()

pandacsv_without_null
```

Mari kita coba validasi lagi dengan cara sebelumnya untuk menampilkan data yang kosong

```python
null_mask = pandacsv_without_null.isnull().any(axis=1)
null_rows = pandacsv_without_null[null_mask]

null_rows
```

### How to fill missing value

```python
pandacsv_fill_na = pd.read_csv('Baby Fashion.csv')

pandacsv_fill_na = pandacsv_fill_na.fillna('Empty')

pandacsv_fill_na
```

Mari kita cek kembali

```python
null_mask = pandacsv_fill_na.isnull().any(axis=1)
null_rows = pandacsv_fill_na[null_mask]

null_rows
```

### Front Fill

```python
pandacsv_ffill_na = pd.read_csv('Baby Fashion.csv')

pandacsv_ffill_na = pandacsv_ffill_na.ffill()

pandacsv_ffill_na
```

```python
null_mask = pandacsv_ffill_na.isnull().any(axis=1)
null_rows = pandacsv_ffill_na[null_mask]

null_rows
```

### Back Fill

```python
pandacsv_bfill_na = pd.read_csv('Baby Fashion.csv')

pandacsv_bfill_na = pandacsv_bfill_na.bfill()

pandacsv_bfill_na
```
