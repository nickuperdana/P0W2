```python
import pandas as pd
```

```python
spotify_df = pd.read_csv('universal_top_spotify_songs.csv')
spotify_df
```

```python
# Cari yang null
spotify_df.isnull()
```

```python
# bisa pakai ini
spotify_df.info()
```

```python
# kalau menampilkan dataset, berarti dia tidak mereplace
spotify_df.dropna()
```

```python
# cara menghapus record yang ada nullnya di satu atau beberapa kolom mana saja

# biar aman, buat salinan database dulu. spotify_clean_df adalah variabel yang isinya data, tidak mereplace file original
spotify_clean_df = spotify_df.copy()

# drop record yang ada null valuenya, inplace=mereplace data
spotify_clean_df.dropna(inplace=True)

# reset index setelah beberapa record null terhapus yang menyebabkan index tidak berurutan dari awal ke akhir
spotify_clean_df.reset_index(inplace=True, drop=True)

# hasil akhirnya, semuanya mengikuti nilai minimum entry yang terisi dengan .info()
spotify_clean_df.info()
```

```python
# kalau pakai subset:
# sepaket dengan yang di list tanpa perlu pikirkan urutannya

spotify_clean_df.drop_duplicates(subset=['name', 'artists'])
```

```python
# ubah tipe data kolom, data convertion

spotify_clean_df.info()
```

```python
# kalau hanya ini, ini hanya menampilkan kolom duration_ms sebagai float
spotify_clean_df.duration_ms.astype('float64')
```

```python
# kalau mau replace kolom duration_ms, kita assign kolom tersebut dengan astype itu.

# float64 adalah tipe data float versi pandas
spotify_clean_df.duration_ms = spotify_clean_df.duration_ms.astype('float64')

spotify_clean_df.info()
```

```python
spotify_clean_df.snapshot_date = spotify_clean_df.snapshot_date.astype('datetime64')

spotify_clean_df.info()
```

```python
# kalau kita lihat, formatnya adalah 'YYYY-MM-DD'
spotify_clean_df.head(10)
```

```python
# null akan menyebabkan error ketika pandas melakukan loop

# str.strip() menghapus spasi yang kelebihan di depan atau di belakang

spotify_clean_df.name = spotify_clean_df.name.str.strip()

spotify_clean_df.head(10)
```

```python
# bikin kolom baru, yang mengambil value dari kolom lain

spotify_clean_df['snapshot_day'] = spotify_clean_df.snapshot_date.dt.day

spotify_clean_df.head(10)
```
